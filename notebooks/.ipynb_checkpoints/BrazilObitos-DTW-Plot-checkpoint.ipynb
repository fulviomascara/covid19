```python
import pandas as pd
from datetime import datetime
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
import numpy as np

today = str(datetime.today().strftime('%Y-%m-%d'))
df = pd.read_csv("../datasets/obito_cartorio.csv")
df = df.filter(["date", "state", "deaths_covid19", "new_deaths_covid19"]).fillna(0)
df = df.sort_values(by=['date'])
df = df.loc[(df['date'] >= '2020-03-01') & (df['date'] <= today)]
sp = df.loc[df['state'] == 'SP']
rj = df.loc[df['state'] == 'RJ']
sp_new = np.array(sp["new_deaths_covid19"])
rj_new = np.array(rj["new_deaths_covid19"])
path = dtw.warping_path(sp_new, rj_new)
dtwvis.plot_warping(sp_new, rj_new, path);
```

```python
states = dict()
for state in df['state']:
    states[state] = np.array(df.loc[df['state'] == state]["new_deaths_covid19"])
```

```python
for A in df['state']:
    for B in df['state']:
        if A == B:
            continue
        path = dtw.warping_path(states[A], states[B])
        dtwvis.plot_warping(s1=states[A], s2=states[B], path=path, filename="{}_{}.jpg".format(A,B));
```
